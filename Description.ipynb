```python
#reading the csv file
dfproducts = pd.read_csv(r"/Users/preciousime/Documents/Python/chatgpt description/description_links_copy.csv", usecols =["link","ChatGPT description"])
dfproducts["length"]=""
dfproducts.tail()
```

```python
#checking for null values
dfproducts.isnull().sum()
```

```python
#sends request to chatgpt
for i in range(0,len(dfproducts)):

    prompt_text = "write a product description for an commerce website focused on selling furniture in 3 sentences ",dfproducts["link"][i]
        
    text = openai.Completion.create(
      model="text-davinci-003",
      prompt=prompt_text,
      max_tokens=150,
      temperature=0.5
    )
    
    dfproducts.loc[i,"ChatGPT description"]=text['choices'][0]['text']
    dfproducts.loc[i,"length"]=len(dfproducts.loc[i,"ChatGPT description"])

dfproducts
```

```python
#remove the \n\n from the result
dfproducts['ChatGPT description'] = dfproducts['ChatGPT description'].str.split('\n\n').str[1]
dfproducts['ChatGPT description']
```
